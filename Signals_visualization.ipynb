```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 2, 1000)  # 2 seconds duration, 1000 samples

# Message signal m(t)
m_t = np.cos(2 * np.pi * t)

# Carrier signal c(t)
c_t = np.cos(20 * np.pi * t)

# AM signal s(t) = m(t) * c(t)
s_t = m_t * c_t

# Plot the signals
plt.figure(figsize=(12, 6))

plt.plot(t, s_t, label='AM Signal s(t)', color='blue')
plt.plot(t, m_t, 'r--', label='Envelope (Message) m(t)')
plt.plot(t, -m_t, 'r--')

plt.title('Amplitude Modulated (AM) Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define frequency axis
f = np.linspace(-15, 15, 1000)

# Define the Fourier Transform S(f) as impulses
S_f = np.zeros_like(f)

# Locations of delta functions (impulses)
impulse_freqs = [-11, -9, 9, 11]
amplitude = 0.25  # each impulse is 1/4

# Add impulses to the plot
for freq in impulse_freqs:
    index = np.argmin(np.abs(f - freq))
    S_f[index] = amplitude

# Plot
plt.figure(figsize=(10, 5))
plt.stem(f, S_f, use_line_collection=True, basefmt=" ")
plt.title('Fourier Transform of $s(t) = \cos(2\pi t) \cdot \cos(20\pi t)$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.ylim(0, 0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define frequency axis
f = np.linspace(-15, 15, 1000)

# Initialize Fourier Transform as all zeros
S_f = np.zeros_like(f)

# Delta function locations and amplitude
impulse_freqs = [-11, -9, 9, 11]
amplitude = 0.25

# Add impulse values to S(f)
for freq in impulse_freqs:
    index = np.argmin(np.abs(f - freq))
    S_f[index] = amplitude

# Plot using basic stem plot
markerline, stemlines, baseline = plt.stem(f, S_f)
plt.setp(markerline, 'markerfacecolor', 'r')
plt.setp(stemlines, 'color', 'b')
plt.setp(baseline, 'color', 'k', 'linewidth', 0.5)

# Labels and formatting
plt.title('Fourier Transform of $s(t) = \cos(2\pi t) \cdot \cos(20\pi t)$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.ylim(0, 0.3)
plt.xlim(-15, 15)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Frequency axis
f = np.linspace(-5, 5, 1000)
pi = np.pi

# Part (a)
sinc_f = np.sinc(f)  # np.sinc(x) = sin(pi x)/(pi x)

# Part (b)
sinc_shifted = np.sinc(f - 1)

# Part (c)
sinc_cos = 0.5 * (np.sinc(f - 1) + np.sinc(f + 1))

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(f, sinc_f)
plt.title('Fourier Transform of p(t)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(f, sinc_shifted)
plt.title('Fourier Transform of $e^{j2\pi t}p(t)$')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(f, sinc_cos)
plt.title('Fourier Transform of $\\cos(2\pi t)p(t)$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(-1, 6, 1000)

# Unit step function
u = lambda x: np.where(x >= 0, 1, 0)

# Unit ramp function
r = lambda x: np.where(x >= 0, x, 0)

# Define the signal
x = r(t) - 2 * u(t - 2) - 2 * u(t - 4)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(t, x, label=r'$x(t) = r(t) - 2u(t-2) - 2u(t-4)$', color='blue', linewidth=2)
plt.title('Plot of $x(t) = r(t) - 2u(t-2) - 2u(t-4)$')
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 1000)

# Original cosine
x1 = np.cos(2 * np.pi * t)

# Phase-shifted cosine
x2 = np.cos(2 * np.pi * t - np.pi / 4)

plt.figure(figsize=(10, 4))
plt.plot(t, x1, label=r'$\cos(2\pi t)$', linestyle='--', color='blue')
plt.plot(t, x2, label=r'$\cos(2\pi t - \frac{\pi}{4})$', color='red', linewidth=2)
plt.title('Phase Shift Comparison')
plt.xlabel('Time (t) [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linestyle='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(-1, 8, 1000)

# Unit step function
u = lambda x: np.where(x >= 0, 1, 0)

# Original signal x(t)
x = u(t) - 2 * u(t - 2) + u(t - 4)

# Transformed signals
x_2t = u(2 * t) - 2 * u(2 * (t - 2)) + u(2 * (t - 4))  # x(2t)
x_t2 = u(t / 2) - 2 * u((t / 2) - 2) + u((t / 2) - 4)  # x(t/2)

# Plot
plt.figure(figsize=(12, 6))

# Plot original x(t)
plt.plot(t, x, label=r'$x(t) = u(t) - 2u(t - 2) + u(t - 4)$', linewidth=2, color='green')

# Plot x(2t)
plt.plot(t, x_2t, label=r'$x(2t)$ (time compression)', linewidth=2, color='blue', linestyle='--')

# Plot x(t/2)
plt.plot(t, x_t2, label=r'$x(t/2)$ (time expansion)', linewidth=2, color='red', linestyle=':')

# Customize plot
plt.title('Plot of $x(t)$, $x(2t)$, and $x(t/2)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(-1, 8, 1000)

# Unit step function
u = lambda x: np.where(x >= 0, 1, 0)

# Original signal x(t)
x = u(t) - 2 * u(t - 2) + u(t - 4)

# Transformed signals
x_2t = u(t / 2) - 2 * u((t / 2) - 2) + u((t / 2) - 4)  # x(2t) (compressed)
x_t2 = u(2 * t) - 2 * u(2 * (t - 2)) + u(2 * (t - 4))  # x(t/2) (expanded)

# Plot
plt.figure(figsize=(12, 6))

# Plot original x(t)
plt.plot(t, x, label=r'$x(t) = u(t) - 2u(t - 2) + u(t - 4)$', linewidth=2, color='green')

# Plot x(2t) (compressed)
plt.plot(t, x_2t, label=r'$x(2t)$ (time compression)', linewidth=2, color='blue', linestyle='--')

# Plot x(t/2) (expanded)
plt.plot(t, x_t2, label=r'$x(t/2)$ (time expansion)', linewidth=2, color='red', linestyle=':')

# Customize plot
plt.title('Plot of $x(t)$, $x(2t)$, and $x(t/2)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.grid(True)
plt.legend()
plt.show()
```
