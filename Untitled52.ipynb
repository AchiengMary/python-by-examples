```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(-1, 5, 1000)

# Unit step function
u = lambda x: np.where(x >= 0, 1, 0)

# Impulse response h(t) = e^(-t) * u(t)
h_t = np.exp(-t) * u(t)

# Output y(t) for x(t) = u(t)
y_t_1 = 1 - np.exp(-t) * u(t)

# Output y(t) for x(t) = u(t) - u(t-2)
y_t_2 = (1 - np.exp(-t)) * u(t) - (1 - np.exp(-(t - 2))) * u(t - 2)

# Plot
plt.figure(figsize=(12, 6))

# Plot h(t)
plt.subplot(3, 1, 1)
plt.plot(t, h_t, label=r'$h(t) = e^{-t} u(t)$', color='green', linewidth=2)
plt.title('Impulse Response $h(t)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')

# Plot y(t) for x(t) = u(t)
plt.subplot(3, 1, 2)
plt.plot(t, y_t_1, label=r'$y(t) = (u(t) * h(t))$', color='blue', linewidth=2)
plt.title('Output $y(t)$ for $x(t) = u(t)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')

# Plot y(t) for x(t) = u(t) - u(t-2)
plt.subplot(3, 1, 3)
plt.plot(t, y_t_2, label=r'$y(t) = (u(t) - u(t-2)) * h(t)$', color='red', linewidth=2)
plt.title('Output $y(t)$ for $x(t) = u(t) - u(t-2)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')

# Show the plots
plt.tight_layout()
plt.show()
```
